#### Use .isin method to filter the dataset and get east asia countries

```python
east_asia = df[df["Country"].isin(["China", "South Korea", "North Korea", "Macau", "Hong Kong", "Tibet", "Taiwan", "Japan", "Mongolia"])]

east_asia
```

### Coding Exercises 4.8

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("population_by_country_2020.csv")

df
```

#### Declare data points

```python
country = east_asia["Country"]
population = east_asia["Population (2020)"]

fig, ax = plt.subplots(figsize=(16, 9))

ax.barh(country,population)

ax.set_title("2020 Population by Country (East Asia)", loc = "center") #title heading of the chart

# remove the axes splines or the box lines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# add annotation in bar population 
for i in ax.patches:
    plt.text(i.get_width() + 0.2, i.get_y() + 0.5, str(round((i.get_width()), 2)), fontsize = 10, fontweight = "bold", color = "grey")

# remove x, y ticks (small black ticks)
ax.xaxis.set_ticks_position("none") 
ax.yaxis.set_ticks_position("none") 

# add spaces between axes and labels
ax.xaxis.set_tick_params(pad=5)
ax.yaxis.set_tick_params(pad=10)

# add gridlines
ax.grid(visible=True, color="grey", linestyle="-.", linewidth=0.5, alpha=0.2)
# plt.grid(axis="x")
# plt.grid(axis="y")

#invert chart
ax.invert_yaxis() 

plt.show()
```

### China has the most highest population in East Asia
