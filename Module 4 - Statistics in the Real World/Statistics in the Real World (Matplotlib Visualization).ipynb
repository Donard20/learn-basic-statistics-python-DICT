get the population statistics from SEA countries

```python
import pandas as pd

df = pd.read_csv("population_by_country_2020.csv")

sea = df[df["Country"].isin(["Philippines", "Thailand", "Indonesia", "Myanmar", "Laos", "Vietnam", "Brunei", "Singapore", "Malaysia", "Cambodia", "Timor Leste"])]

sort_sea = sea.sort_values(by = "Population (2020)", ascending = False)

sort_sea
```

By sorting the datafield of sea countries, we can conclude that "Indonesia" had the highest share in population among the asean contries

### MATPLOTLIB VISUALZATION

```python
import matplotlib.pyplot as plt  #importatnt: import first the module

sea = df[df["Country"].isin(["Philippines", "Thailand", "Indonesia", "Myanmar", "Laos", "Vietnam", "Brunei", "Singapore", "Malaysia", "Cambodia", "Timor Leste"])]

country = sea["Country"]
population = sea["Population (2020)"]

fig = plt.figure(figsize = (16, 9)) #size of the chart

plt.title("2020 Population by Country (SEA)", loc = "center") #title for the chart

plt.barh(country, population) #.barh horizontal barchart x-axis(country) y-axis(population)

plt.show #display
```

## SEA Population

```python
country = sea["Country"]
population = sea["Population (2020)"]

fig, ax = plt.subplots(figsize=(16, 9)) #size of the chart

ax.barh(country,population) #axes of chart

ax.set_title("2020 Population by Country (SEA)", loc = "center") #title heading of the chart

# remove the axes splines or the box lines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# add annotation in bar population 
for i in ax.patches:
    plt.text(i.get_width() + 0.2, i.get_y() + 0.5, str(round((i.get_width()), 2)), fontsize = 10, fontweight = "bold", color = "grey")

# remove x, y ticks (small black ticks)
ax.xaxis.set_ticks_position("none") 
ax.yaxis.set_ticks_position("none") 

# add spaces between axes and labels
ax.xaxis.set_tick_params(pad=5)
ax.yaxis.set_tick_params(pad=10)

# add gridlines
ax.grid(visible=True, color="grey", linestyle="-.", linewidth=0.5, alpha=0.2)
# plt.grid(axis="x")
# plt.grid(axis="y")

#invert chart
ax.invert_yaxis() 

plt.show()
```

#### As we can see in the chart, Indonesia had the most highest populated country among the SEA countries in 2020 and Brunei is the lowest population.
