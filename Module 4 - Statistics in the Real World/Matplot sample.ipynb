```python
import pandas as pd
```

```python
f_usage = [5, 4, 3, 5, 8, 2, 5, 2, 5, 8, 10, 8, 7, 9, 10, 5, 7, 5, 7, 5, 4, 3, 5, 8, 4]

ser = pd.Series(f_usage)
sum_stat = ser.describe()


print(sum_stat)
```

## Visualization

```python
import matplotlib.pyplot as plt  #--display chart
import numpy as np #--statistics library
import statistics as stat


fig = plt.figure(figsize = (16, 9)) # 2-- resizing the chart

plt.hist(f_usage, bins = 8, edgecolor = "black") # 1---create barchart histogram with 8 bins

plt.xlabel('Hours')  #3--label the x-axis
plt.ylabel('Frequency') #4 --label the y-axis
plt.title('Friends Social Media Usage', fontsize = 14) #5--display title for the chart 

#Applying Measures of Central Tendency and Dispersion
plt.plot([np.mean(f_usage), np.mean(f_usage)], [0, 9], linestyle = "--", color = "r")
plt.plot([np.median(f_usage), np.median(f_usage)], [0, 9], linestyle = "--", color = "y")
plt.plot([stat.pstdev(f_usage), stat.pstdev(f_usage)], [0, 9], linestyle = "--", color = "g")
plt.plot([stat.mode(f_usage), stat.mode(f_usage)], [0, 9], linestyle = "--", color = "m")
plt.plot([stat.variance(f_usage), stat.variance(f_usage)], [0, 9], linestyle = "--", color = "c")

#display the color history
plt.text(8, 7, 'mean', color = "r")
plt.text(8, 6.5, 'median', color = "y")
plt.text(8, 6, 'mode', color = "m")
plt.text(8, 5.5, 'std', color = "g")
plt.text(8, 5, 'variance', color = "c")

plt.show #--display the chart
```
